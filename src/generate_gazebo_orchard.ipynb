```python
n_rows = 3
n_cols = 10
n_bushes = 3
n_trees = n_rows * n_cols * n_bushes
spacing = 0.75
bush_spacing = 3.75
```

```python

cnt = 0
for row in range(int(-n_rows/2),int(n_rows/2)):
    for col in range(int(-n_cols/2),int(n_cols/2)):
        for bush in range(n_bushes):
            i = cnt
            roll = np.random.normal(0,0.06)
            pitch = np.random.normal(0,0.06)
            yaw = np.random.uniform(0,np.pi*2)
            jitter_x = np.random.uniform(-0.15,0.15)
            jitter_y = np.random.uniform(-0.15,0.15)
            state.add_model('asparagus_{}'.format(i))
            state.models[i].pose = [row*spacing + jitter_x + (bush-1) * bush_spacing, col*spacing + jitter_y,0,roll,pitch,yaw]
            state.models[i].add_link('link')
            state.models[i].links[0].pose = [row*spacing + jitter_x + (bush-1) * bush_spacing, col*spacing + jitter_y,0,roll,pitch,yaw]
            cnt += 1
```

```python
for i in range(n_trees):
    scale = 0.08
    #scale_z = np.random.uniform(0.0125,0.15)
    scale_z = np.random.normal(0.08,0.015)
    asparagus = create_sdf_element('model')
    asparagus.static = True
    asparagus.name = "asparagus_{}".format(i)
    asparagus.add_link(name="asparagus_{}".format(i))
    visual = create_sdf_element('visual')
    visual.geometry.mesh = create_sdf_element('mesh')
    visual.geometry.mesh.uri="model://Asparagus/meshes/asparagus.obj"
    visual.geometry.mesh.scale=[scale, scale, scale_z]
    collision = create_sdf_element('collision')
    collision.geometry.mesh = create_sdf_element('mesh')
    collision.geometry.mesh.uri="model://Asparagus/meshes/asparagus.obj"
    collision.geometry.mesh.scale = [scale, scale, scale_z]
    collision.max_contacts = 10
    asparagus.links[0].self_collide = False
    asparagus.links[0].enable_wind = False
    asparagus.links[0].kinematic = False
    asparagus.links[0].add_visual("visual",visual)
    asparagus.links[0].add_collision("collision",collision)
    world.add_model('asparagus_{}'.format(i), asparagus)
```

```python
world.add_state("default",state)
```
